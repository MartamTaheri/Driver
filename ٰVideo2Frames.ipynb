# Convertor function

```python
def convertor(video_path, skip=56):
  print(f"{video_path} Started.")  
  vidcap = cv2.VideoCapture(base_url + video_path)
  success,image = vidcap.read()
  count = 0    
  while success:    
    if(count%skip == 0):
      cv2.imwrite(f"{store_in}{video_path}{count}.jpg", image)     # save frame as JPEG file                
    success,image = vidcap.read()        
    count += 1
  print(f"{video_path} Finished.")
```

```python
for f in files:
  convertor(f)
```
