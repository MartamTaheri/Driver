# Intensity Transformation:

**negative Transform :** converting the black regions to white and vice versa





```python
def negative_image(url):
    img = io.imread(url )
    negative_image = 255 - img

    im_c = cv.hconcat([img, negative_image])
    print()
    cv.imwrite('data/dst/opencv_vconcat.jpg', im_c)
    cv2_imshow(im_c)
    
    cv.waitKey(0)
    cv.destroyAllWindows()
    return
```

**log transform**

```python
def log_image (url):

    img = io.imread(url)
  
    c = 255 / np.log(1+ np.max(img))
    log_image = c *(np.log(img + 1))
    log_image = np.array(log_image, dtype = np.uint8)
    im_c = cv.hconcat([img, log_image])
    cv.imwrite('data/dst/opencv_vconcat.jpg', im_c)
    print()
    cv2_imshow(im_c)

    cv.waitKey(0)
    cv.destroyAllWindows()

    return
```

```python

imgs_url= ['/content/drive/MyDrive/some_pictures/images.jfif',
           '/content/drive/MyDrive/some_pictures/download.jfif'
                      
  ]
for url in imgs_url:
  img = io.imread(url )    

  cv.imwrite('image.jpg', img)
  log_image(url)
```

**power_law (gamma) Transform **


```python

def gamma_img(url):
    
    img = io.imread(url)

    for gamma in [0.1, 0.5, 1, 1.5, 2, 2.5]:
        power_law= np.array(255 * ((img / 255)**gamma), dtype = 'uint8' )
        cv.imwrite('gamma_transformed'+str(gamma)+'.jpg', power_law)


    # im_c = cv.hconcat([img, power_law])
    # cv.imwrite('data/dst/opencv_vconcat.jpg', im_c)
    # print()
    # cv2_imshow(im_c)
    

    cv.waitKey(0)
    cv.destroyAllWindows()

    return
```

```python

imgs_url= ['/content/drive/MyDrive/some_pictures/images.jfif',
           '/content/drive/MyDrive/some_pictures/download.jfif'            
  ]
for url in imgs_url:
  img = io.imread(url )

    

  gamma_img = gamma_img(img)

  fig, axes = plt.subplots(1, 2, figsize=(15, 10))
 

  ax[0].imshow(img)
  ax[0].set_title("Original")
  ax[1].imshow(gamma_img)
  ax[1].set_title("gamma img")

  
  
  fig.tight_layout()

  plt.show()
```
